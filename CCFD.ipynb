```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,confusion_matrix,classification_report,accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("/content/creditcard.csv")
df
```

```python
df.describe()
```

```python
x_val=df.drop("Class",axis=1)
y_val=df["Class"]
```

```python
non_nan_mask = y_val.notna()
x_val_cleaned = x_val[non_nan_mask]
y_val_cleaned = y_val[non_nan_mask]
x_train,x_test,y_train,y_test=train_test_split(x_val_cleaned,y_val_cleaned,test_size=0.2,random_state=42,stratify=y_val_cleaned)
```

```python
model=LogisticRegression(max_iter=1000,class_weight="balanced")
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print("Model prediction :\n",pd.Series(y_pred).value_counts())
```

```python
acc=accuracy_score(y_test,y_pred)
print("Accuracy :",acc)
```

```python
cm=confusion_matrix(y_test,y_pred,labels=[0,1])
print("Confusion Matrix :\n",cm)
```

```python
cs=classification_report(y_test,y_pred)
print("Classification Report :\n",cs)
```

```python
mean_ae=mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error :",mean_ae)
mean_se=mean_squared_error(y_test,y_pred)
print("Mean Squared Error :",mean_se)
```
